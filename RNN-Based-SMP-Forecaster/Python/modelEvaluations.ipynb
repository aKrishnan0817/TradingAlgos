```python
spyData=pd.read_csv("spydata.csv",index_col=0)
```

```python
training_data_len = math.ceil(len(spyData) * .8)
training_data_len

#Splitting the dataset
train = spyData[:training_data_len].iloc[:,3:4]
test = spyData[training_data_len:].iloc[:,3:4]
print(train.shape, test.shape)
```

```python
pred= pd.read_csv("rnnPrediction-1.csv")
y_pred_RNN = pred.loc[:,'Close']
```

```python
pred= pd.read_csv("lstmPrediction-1.csv")
y_pred_GRU = pred.loc[:,'Close']
```

```python
(y_pred_GRU[0]+y_pred_RNN[0])/2
```

```python
#GruRNNmean= pd.DataFrame([y_pred_GRU,y_pred_RNN]).mean()
GruRNNmean= list()
for i in range(len(y_pred_GRU)):
  GruRNNmean.append((y_pred_GRU[i]+y_pred_RNN[i])/2)
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(
    x=pd.to_datetime(test.index),
    y=test["Close"],
    mode='lines',
    name='Test Data'
))
fig.update_layout(title="Model Predictions Simple RNN")
fig.add_trace(go.Scatter(
    x=pd.to_datetime(train.index),
    y=train["Close"],
    mode='lines',
    name='Train Data'
))
fig.add_trace(go.Scatter(
    x=pd.to_datetime(test.index),
    y=y_pred_RNN,
    mode='lines',
    name='RNN prediction'
))
fig.add_trace(go.Scatter(
    x=pd.to_datetime(test.index),
    y=y_pred_GRU,
    mode='lines',
    name='GRU prediciton'
))
fig.add_trace(go.Scatter(
    x=pd.to_datetime(test.index),
    y=GruRNNmean,
    mode='lines',
    name='GRU RNN mean'
))
```
