```python
import pandas as pd
import pandas_market_calendars as mcal
import datetime
import yfinance as yf
import pickle
```

```python
def get_calendar():
    nyse = mcal.get_calendar('NYSE')
    a = nyse.valid_days(start_date='2014-01-01', end_date='2024-11-08')
    cal =[]
    for x in a:
        cal.append(datetime.datetime.strftime(x,'%Y-%m-%d'))
    return cal
```

```python
def getDateOffset(Date,offset=0):
    try:
        return cal[cal.index(Date)+offset]
    except:
        cal=get_calendar()
        return cal[cal.index(Date)+offset]
```

```python
def getDate(date):
    #date needs to be a string like 2024-01-01
    dateList= date.split('-')
    return datetime.date(int(dateList[0]),int(dateList[1]),int(dateList[2]) )
```

```python
def formatData(year):
    data = pd.read_csv(f"RawData/NDXPriceData{year}.csv")
    for i in range(len(data)):
        data["Ticker"][i] = data["Ticker"][i].split(" ")[0]

    Tickers = data["Ticker"].drop_duplicates()
    dates = list(data["Date"].drop_duplicates())
    dataFormated = pd.DataFrame({"Date":dates}).set_index("Date")

    for ticker in Tickers:
        d = data[data['Ticker'] == ticker].set_index("Date")
        priceList=[None]*len(dates)
        firstIndex = dates.index(d.index[0])
        for i in dates[firstIndex:]:
            try:
                priceInfo=({"Open":d.loc[i,"Open"],"Close":d.loc[i,"Close"],"High":d.loc[i,"High"],"Low":d.loc[i,"Low"]})   

            except:
                try:
                    d1 = (yf.download(ticker, start=i, end=getDateOffset(i,1),multi_level_index=False)).reset_index().set_index("Date")
                    j= d1.index[0] 
                    priceInfo=({"Open":d1.loc[j,"Open"],"Close":d1.loc[j,"Close"],"High":d1.loc[j,"High"],"Low":d1.loc[j,"Low"]}) 
                except:
                    priceInfo = None
            priceList[dates.index(i)]=priceInfo

        dataFormated[ticker] = priceList

    return dataFormated
```

```python
year=2019
```

```python
newData = formatData(year)
newData.to_pickle(f"../Stock_Picker_Algo/Tools/NDX_data/NDXdata{year}.pkl")
```
