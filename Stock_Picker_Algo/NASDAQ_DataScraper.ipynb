# Data Collection
    Using a framework that permits us to gather the tickers of the NASDAQ index over the years, I compiled a list of all the tickers over the past 5 years.
    
    After creating the list used yahoo finance to gather each stocks data over the same time interval

```python
startYear=2020
endYear=2025
```

```python
def flat(lis):
    flatList = []
    # Iterate with outer list
    for element in lis:
        if type(element) is list:
            # Check if type is list than iterate through the sublist
            for item in element:
                flatList.append(item)
        else:
            flatList.append(element)
    return flatList
```

```python
def getTickers(startYear=2020, endYear=2025):
    tickers=[]
    for year in range(startYear,endYear):
        tickers.append(list(ndx.tickers_as_of(year)))
    arr = np.array(flat(tickers))
    return np.unique(arr)
```

```python
tickers = getTickers(startYear,endYear)
```

```python
cal= get_calendar(str(startYear))
startDate = cal[0]
endDate=cal[-1]
```

```python
tickers=list(tickers)+["ARM","LIN","SMCI"]
```

# Data Consolidation
    At the moment all the data is in ~130 csv files, the goal is to consolidate it into a single data set 
    The data set should include every stocks and for each stock the closing and opening price for each day

```python
joinedData = pd.DataFrame()
```

```python
for ticker in tickers:
    try:
        with open('NDX_data/'+ticker+'.pkl', 'rb') as file:
            data = pickle.load(file)  
            
        dataList = ["N/A"]*1222
        index_2= len(data)-1
        for i in range(len(dataList)-1, 0, -1):
            try:
                dataList[i] = ({"Close":data["Close"][index_2],"Open":data["Open"][index_2]})
            except:
                break
            index_2 -=1    

        joinedData[ticker]=dataList
    except:
        print(ticker," data not available")
```
